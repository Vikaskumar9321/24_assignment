#Ans. 5

1.) 'DataFrame.size':

.The 'DataFrame.size' attribute returns the total number  of  elements in the DataFrame. It represents the number of rows multiplied by the number of columns in the DataFrame.

.The returned value is an integer.

.it is calculated as 'DataFrame.Shape[0]' * 'DataFrame.size[1]'.
for example if a DataFrame has 5 rows and 3 columns, the size attribute will be 15.


2.) 'DataFrame.shape':


.The 'DataFrame.shape' attribute returns a tuple that represents the dimensions of the DataFrame.

.The tuple contains two values: the the number of rows and the number of columns respectively.

.The returned value is a tuple of integers.
for example, if a DataFram has 5 rows and 3 columns , the shape attribute will be(5,3)

```python
def select_rows(df8):
    selected_rows = df8[(df8['A'] > 5) & (df8['B'] < 10)]
    return selected_rows
```

```python
data = {'A' : [3,8,6,2,9],
        'B' : [5,2,9,3,1],
        'C' : [1,7,4,5,2]}
df8 = pd.DataFrame(data)

df = select_rows(df8)
```

#Ans.9

```python
def calculate_stats(df):
    mean_value = df['Values'].mean()
    median_value = df['Values'].median()
    std_value = df['Values'].std()
    
    return mean_value, median_value, std_value
```

```python
data = {'Values' : [2,4,6,8,10]}
df = pd.DataFrame(data)
mean, median, std = calculate_stats(df)
```

```python
"Mean:", mean
```

```python
"Median:", median
```

```python
"Standard Deviation:", std
```
